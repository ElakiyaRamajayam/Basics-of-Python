***Calculator***

```python
a = int(input("Enter the value for a: "))
b = int(input("Enter the value for b: "))
operator = input("Enter the operator: ")
if operator == "+":
  print(a+b)
elif operator == "-":
  print(a-b)
elif operator == "*":
  print(a*b)
elif operator == "/":
  print(a/b)
elif operator == "%":
  print(a%b)
else:
  print("Error")

```

***Even sum***

```python
even_sum = 0
lst = [1, 3, 4, 6, 8, 3, 8, 9]
for x in lst:
  if x%2 == 0:
    even_sum = even_sum + x
print(even_sum)
```

***Vowel count***

```python
vowel_count = 0
vowel = "aeiouAEIOU"
word = "elakiya"
for ch in word:
  if ch in vowel:
    vowel_count += 1
print(vowel_count)
```

***Random Password***

```python
import random

chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"

password = ""

for i in range(8):
    password += random.choice(chars)

print(password)
```
